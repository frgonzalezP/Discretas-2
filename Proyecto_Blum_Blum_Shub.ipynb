##Introducción
La seguridad de información dentro de los ambientes laborales actuales resulta ser un problema, de esta manera, al momento de enviar o transportar información, esta puede ser robada o filtrada, exponiendo así, posible información sensible o afectado negocios etc.
De esta manera la aplicación de contraseñas para asegurar los dispositivos que almacenan información o los propios archivos, es fundamental actualmente, de esta manera, la propuesta expuesta en este trabajo es la utilización de el algoritmo BBS o Blum Blum Shub, como solución rápida y simple, para generar contraseñas seguras, compuestas exclusivamente de números, además de tener la capacidad de ser compartidas a 3ros involucrados de manera segura, es decir, se puede transmitir la contraseña sin necesidad de que se descubra el método de generarla.
## Estado de arte
El Blum Blum Shub como algoritmo de generación de números pseudoaleatorios, surge en el 86, como respuesta a la búsqueda de un método para generar cadenas de bits Pseudoaleatorios que no fueran predecibles, para de esta manera, evitar depender de métodos generadores de números aleatorios tales como ERNIE 1 o el listado de números generados por la RAND Corporation.


Por su parte ERNIE era un generador de números aleatorios que usaba hardware para obtenerlos, su primera versión introducida en 1957, generaba números aleatorios a partir de la señal de ruido generada por tubos de neón, por su parte RAND Corporation género en 1955, una tabla que contenía 1 millon de números aleatorios, generados usando una Ruleta electrónica, esto dada la necesidad de números aleatorios para la época.


Desde la década de los 70, la necesidad de tener métodos rápidos y reproducibles, para números aleatorios, da lugar a los algoritmos enfocados en números Pseudoaleatorios (Por su propiedad de ser reproducibles), aunque estos presentaban problemas, tales como su predictibilidad o su dificultad de procesamiento para la época.


Actualmente, La generación de cadenas de números aleatorios se lleva a cabo usando en algunos casos, Métodos Cuánticos, este es el caso de ERNIE 5, que usando luz, supera con creses la velocidad de la primera generación, realizando el trabajo de generar los ganadores de lotería, en este caso, al ser bonos emitidos por el gobierno, esta lotería debería ser lo mas justa posible, La primer generación, tardaba 3 días en generar los números ganadores, por el contrario, la nueva generación, demora 12 minutos.


Para La generación de números Pseudoaleatorios, antes se usaba el método de la parte media del cuadrado, propuesto por Jon Von Neuman en 1946, pero este método tenia fallas, puesto que en ciclos cortos podía o repetirse o volverse “Monótono” gracias a que, si existía un cero este se propagaría por siempre dentro de los cálculos.

Es decir, r es residuo cuadratico modulo m, pues es el residuo que deja la division de un cuadrado perfecto entre m,si y solo si, existe un X que al ser elevado al cuadrado, cumple la propiedad, ademas cabe aclarar que m es primo.

##Funcionamiento del BBS

El BBS funciona usando los siguientes pasos

1. Se generan 2 numeros Blum, muy grandes, llamados p y q
2. se multiplican para generar M, este m servira como nuestro modulo
3. se escoje la semilla, en el intervalo (1,m-1), este valor sera denominado x sub 0
4. se aplica la formula Formula 10.JPG


Este procedimiento dara como resultado un numero aleatorio cada vez que sea aplicado, cabe destacar, que se trata de un metodo recursivo, pues depende de los vaores anteriores, para generar el siguiente.

##Seguridad y resultado del generador

Teniedo en cuenta que, Si el valor de X o el valor de M se llegase a conocer, se facilitaria la ruptura del codigo, por lo tanto dejara de ser efectivo,de esta manera, no podemos solo generar los numeros, debemos usarlos para crear un codigo que no aporte informacion que permita este tipo de casos.

Debido a este problema, los creadores del generador tuvieron una buena idea, que tal si, en lugar de entregar el numero completo, se muestra solo 1 bit de los que componen al numero, de ser asi, la eleccion del bit , se podria pensar que es aleatorio, pero no, el objetivo es dar la menor cantidad de informacion posible, pero que permita al generador ser generar resultados iguales si los valores de m y x son correctos.

Para resolver este problema, se decidio usar el bit indicador de paridad, es decir, usar la operacion modulo 2, para generar valores de 0 y 1, a pesar de lo que se pueda pensar, siguiendo las instrucciones y teniendo en cuenta no rebelar mas informacion de la necesaria, el BBS es un generador de numeros confiable.

````
# Experimento numero 1
import random
from time import time
def procedimiento(lista, modulo):
  for i in range (15000):
    resultado=((lista[i]**2)%modulo)
    lista.append(resultado)
 

def output(salida,lista):
  for numero in lista:
    resul=numero%2
    if resul!=0:
      salida.append(1)
    elif resul==0:
      salida.append(0)

timeinicio=time()
p=7
q=19
modulo=p*q
semilla_pass=0
while (semilla_pass==0):
  seed = random.randint(1,modulo)
  if seed!=p and seed !=q:
    semilla_pass=1


lista=[]
salida=[]
lista.append(seed)
procedimiento(lista,modulo)
output(salida,lista)
print(salida)
print(lista)
print("Semilla ", seed, " numero P ",p," Numero q ",q," modulo ",modulo)
timefinal=time()-timeinicio
print("tiempo de ejecucion ",timefinal)
````

````
# Experimento numero 4

import sympy
import random
import math 


def primo_compatible(numero):
  primo=sympy.nextprime(numero)
  while (primo % 4 != 3):
    primo = sympy.nextprime(primo)
  return primo

def procedimiento(lista, modulo):
  for i in range (35000):
    resultado=((lista[i]**2)%modulo)
    lista.append(resultado)


def output(salida,lista):
  for numero in lista:
    resul=numero%2
    if resul!=0:
      salida.append(1)
    elif resul==0:
      salida.append(0)


aleatorio=random.randint(1,9)
n=aleatorio*10**10
m=(aleatorio+1)*10**10

p=primo_compatible(n)
q=primo_compatible(m)
modulo=p*q

semilla_pass=0
while (semilla_pass==0):
  seed = random.randint(1,modulo)
  if seed!=p and seed !=q:
    semilla_pass=1

lista=[]
salida=[]
lista.append(seed)
procedimiento(lista,modulo)
output(salida,lista)

Indice_inicial_contraseña=random.randint(1,len(salida)-80)

contraseña_binario=[]
for i in range(80):
  contraseña_binario.append(salida[Indice_inicial_contraseña+i])


valores_invididuales=[]
for i in range(0,len(contraseña_binario)):
  valores_invididuales.append(int(contraseña_binario[i])*2**i)

suma=0
for i in range(0,len(valores_invididuales)):
  suma+=valores_invididuales[i]


print("Su contraseña es ",suma)
````
